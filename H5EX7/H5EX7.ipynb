### (b) Using Python, plot the values of $|X (e^{jω})|$ over a set of uniformly spaced frequency values between $−π$ and $π$; set $M = 20$ and use 10,000 frequency values

```python
M = 20
freq_val = 10_000
w = np.linspace(-np.pi,np.pi,freq_val)
X = np.abs(np.sin(w*M/2)/np.sin(w/2))

plt.plot(w,X)
plt.xlabel('w')
plt.ylabel('$|X(e^{jw})|$')
plt.title('Magnitude of $X(e^{jw})$')
plt.show()
```

### (c) Now generate an $N$-point finite-length sequence $x_N [n]$ where $x_N [n] = x [n]$ for $n = 0, 1, . . ., N − 1$. Compute its DFT using NumPy’s FFT package and plot the magnitude of the coefficients for $N = 31, 51, 101$. Align the DFT plots so that you can visually compare the DFT coefficients to the analytical values of the DTFT obtained in the previous step

```python
N =[31,51,101]
for n in N:
    # create signal x
    x = np.zeros(n)
    x[:M] = 1
    # Compute FFT and shift it so that the zero frequency is in the middle
    X_fft  = np.fft.fft(x)
    X_fft_shifted = np.fft.fftshift(X_fft)
    # map the frequency values to the range [-pi,pi] to be able to compare with the analytical expression
    w = np.linspace(-np.pi,np.pi,n)
    plt.plot(w,np.abs(X_fft_shifted))
    plt.xlabel('w')
    plt.ylabel('$|X(e^{jw})|$')
    plt.title('Magnitude of $X(e^{jw})$ for N = ' + str(n))
    plt.show()
```
